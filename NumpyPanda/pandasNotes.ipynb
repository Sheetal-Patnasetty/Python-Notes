# Handling Nulls

```python
import pandas as pd
```

```python
nba = pd.read_csv('nba.csv')
```

```python
nba.head()
```

```python
nba.isna().sum()
```

### There are 2 methods using which you can handle Nulls

Method-1: you can decide to delete/ drop rows containing null using dropna() method.

Method-2: you can decide to replace or impute nulls using fillna() method. 

## Method-1:

### dropna ( how = ' all | any ' , inplace = True )

inplace = True -> if you want to make changes in the original dataFrame

dropna(how='any', inplace=True) -> how = 'any' means delete a row only if any column contains NaN.

dropna(how='all', inplace=True) -> how = 'all' means delete a row only if all columns contain NaN.

```python
len(nba)
```

```python
nba.dropna(how='all', inplace=True)
```

```python
len(nba)
```

```python
nba.dropna(how='any', inplace=True)
```

```python
len(nba)
```

=> 93 rows deleted containing a null in any of their columns

```python
nba.isna().sum()
```

## Method-2: Impute rows using fillna() method

```python
# Reload the dataset 

nba = pd.read_csv('nba.csv')
```

```python
nba.head()
```

```python
# Average of salary
round(nba.Salary.mean(),2)
```

```python
# Fill the Salary column nulls with the column's average.
nba.Salary.fillna(round(nba.Salary.mean(),2), inplace=True)
```

```python
nba.head()
```

```python
# Fill 'Others' in the college column's null spaces. 

nba.College.fillna('Other', inplace=True)
```

```python
nba.College.value_counts()
```

```python
len(nba[nba.College=='Other'])
```

# Pivot

- for reshaping and better readability of data.

```python
import pandas as pd
```

```python
# prepare a sample data

data = {
    'Region' : ['North','South','East','West','North','South','East','West'],
    'Product': ['A','A','A','A','B','B','B','B'],
    'Month'  : ['Jan','Jan','Jan','Jan','Feb','Feb','Feb','Feb'],
    'Sales'  : [250, 150,200,100,300,400,350,200]
}

df = pd.DataFrame( data )
```

```python
df
```

```python
# Monthly sales in all regions
pivot_df = df.pivot (index='Month', columns='Region', values='Sales')
```

```python
pivot_df
```

```python
# Average sales in a month
pivot_df = df.pivot_table(columns='Month', values='Sales', aggfunc='mean')
```

```python
pivot_df
```

```python
yt.head(1)
```

```python
yt.shape
```

```python
# How many channels are listed in Data:

yt.channel_title.nunique()
```

```python
# check nulls:

yt.isna().sum()
```

```python
# Number of duplicates = number of rows - number of unique values 
# keep video_id as index

yt.shape[0] - yt.video_id.nunique()
```

```python
# delete duplicates
yt.drop_duplicates(subset=['video_id'], keep='last', inplace=True)
```

```python
len(yt)
```

```python
# Check datatypes and change the datatypes of those required columns

yt.dtypes
```

```python
# Change object -> datetime

yt['publish_time'] = pd.to_datetime(yt['publish_time'])
```

```python
yt.dtypes
```

```python
# Change object -> datetime

yt['trending_date']=pd.to_datetime(yt['trending_date'],format='%y.%d.%m') 
```

```python
yt.dtypes
```

```python
# Find the most viewed video uploaded by T-Series in the year 2017

yt_sorted = yt[(yt.channel_title == 'T-Series') & (yt.publish_time.dt.year == 2017)].sort_values(by='views', ascending=False)
yt_sorted.head(1)
```

```python
# Number of videos for every month

yt.pivot_table(index=yt.publish_time.dt.month, values='video_id', aggfunc='count')
```

```python
# Find the month which has more number of views.

yt.groupby(yt['publish_time'].dt.month).agg({'views':'sum'}).sort_values(by='views', ascending=False).head(1)
```
